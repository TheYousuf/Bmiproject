```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Series
Section one is about series in python

```python
a = pd.Series(range(1,10))
b = pd.Series(["I", "like", "to", "use", "python", "and", "pandas", "very", "much"], index=range(0,9))
print(a,b)
```

```python
start_date = "Jan 1, 2020"
end_date = "Dec 31, 2020"
idx = pd.date_range(start_date, end_date)
print(f"Length of index is (len(idx))")

items_sold = pd.Series(np.random.randint(25, 50, size=len(idx)), index=idx)
items_sold.plot(figsize=(8, 5)) 
plt.show()
```

# DataFrames
These are data in a tabular format

```python
a = pd.Series(range(1,10))
b = pd.Series(["I", "like", "to", "use", "python", "and", "pandas", "very", "much"], index=range(0,9))
df = pd.DataFrame([a,b])
df
```

```python
df.T
```

```python
df.T.rename(columns={0:"A", 1:"B"})
```

```python
df = df.T.rename(columns={0:"A", 1:"B"})
df
```

```python
df
```

```python
df[(df["A"]>5) & (df["A"]<7)]
```

```python
df["DivA"] = df["A"] - df["A"].mean()
df
```

```python
df["LenB"] = len(df["B"])
df
```

```python
df["LenB"] = df["B"].apply(len)
df
```
