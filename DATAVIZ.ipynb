```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/microsoft/Data-Science-For-Beginners/main/data/birds.csv")
print(df.to_string())
```

```python
df.head()
```

```python
df["MaxWingspan"].plot()
plt.show()
```

```python
plt.title('Max Wingspan in Centimetres')
plt.ylabel('WIngspan (CW)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = df['Name'] 
y = df['MaxWingspan']

plt.plot(x,y)
plt.show()
```

```python
df.nunique()
```

```python
plt.title('Max Wingspan in Centimetres')
plt.ylabel('WIngspan (CW)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(df)):
    x = df['Name'][i]
    y = df['MaxWingspan'][i]
    plt.plot(x,y, 'bo')
    if df['MaxWingspan'][i] > 500:
        plt.text(x,y *(1 - 0.05), df['Name'][i], fontsize=12)

plt.show()
```

```python
plt.title('Max Wingspan in Centimetres')
plt.ylabel('WIngspan (CW)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
for i in range(len(df)):
    x = df['Name'][i]
    y = df['MaxWingspan'][i]
    if df['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, 'bo')
plt.show()
```
